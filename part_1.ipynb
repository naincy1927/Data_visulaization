```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0,10) 
y = 2*x
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0,10) 
y = 2*x
plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("this is mine first plot ")
plt.xlim(0,7)
plt.ylim(0,12)
plt.show()
```

```python
a = np.linspace(0,10,11)
print(a)
b = a**4
print(b)
f = plt.figure()
axes = f.add_axes([0,0,1,1]) ## there are 4 types of axies that is left , bottom , widht , height
axes.plot(x,y)
plt.show()
```

```python
f = plt.figure()
axes = f.add_axes([0,0,1,1]) ## there are 4 types of axies that is left , bottom , widht , height
axes.plot(a,b)
plt.show()
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.2,0.5,0.5])
axes1.plot(a,b)
axes1.set_xlabel("mine axies")
axes1.set_ylabel("mine axies 2")
axes2.plot(a,b)
plt.show()
```

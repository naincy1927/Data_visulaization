```python
import matplotlib.pyplot as plt
import numpy as np 
x = np.arange(0,10)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2)
ax.plot(x,x**3)
```

```python
x = np.arange(0,10)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label= 'Line - 1')
ax.plot(x,x**3,label='Line - 2')
ax.legend()
```

```python
x = np.arange(0,10)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label= 'Line - 1')
ax.plot(x,x**3,label='Line - 2')
ax.legend(loc=1) ##vale of loc goes from 0 to 4 it shifts this small box in the image
```

```python
import matplotlib.pyplot as plt
import numpy as np
a = np.linspace(0,10)
b = a**4
x = np.arange(0,10)
print(x)
y = 2*x
print(y)
 
```

```python
fig,axes = plt.subplots()
axes.plot(x,y,"r")
```

```python
fig , axes = plt.subplots(nrows= 1,ncols= 3)
```

```python
fig,axes = plt.subplots(nrows= 1,ncols=2)
for axe in axes:
    axe.plot(x,y,"r")
```

```python
fig,axes = plt.subplots(nrows= 1,ncols=2)
axes[0].plot(a,b,"m")
axes[1].plot(x,y,"k")
```

```python
axes[0][0].plot(a,b)
axes[1][1].plot(x,y)
```

```python
fig,axes = plt.subplots(nrows= 2,ncols=2)

axes[0][0].plot(a,b)
axes[1][1].plot(x,y)
axes[0][1].plot(a,b)
axes[1][0].plot(x,y)
```

```python
fig,axes = plt.subplots(nrows= 2,ncols=2)
axes[0][0].plot(a,b)
axes[1][1].plot(x,y)
plt.tight_layout()
```
